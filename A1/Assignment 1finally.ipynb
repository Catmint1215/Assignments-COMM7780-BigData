```python
import requests 
import csv
import re
from bs4 import BeautifulSoup
a="https://cd.lianjia.com/ershoufang/pg"
b= 1 
c="co32/"
url=a+str(b)+c

items = [None]*10*30
index=0
while b<=10:
    r = requests.get(url).text
    data = BeautifulSoup(r,"html.parser")

    num = data.findAll("div",attrs = {"class":"address"})
    i=0
    ii=len(num)

    while i<ii:
        m=0
        item=[None]*5
        #print("\n",i)
        #===========================================================

        what = data.findAll("div",attrs = {"class":"title"})

        title = re.search('blank">(.*)</a>',str(what[i+1])).group(1)

        #print(tittle)

        #============================================================

        what = data.findAll("div",attrs = {"class":"address"})

        name = re.search('blank">(.*)</a>',str(what[i])).group(1)

        #print(name)

        #============================================================
        what = data.findAll("div",attrs = {"class":"flood"})

        district = re.search('blank">(.*)</a>',str(what[i])).group(1)

        #print(district)

        #============================================================

        what = data.findAll("div",attrs = {"class":"priceInfo"})

        unitprice = re.search('<span>单(.*)</span>',str(what[i])).group(1)

        #print(unitprice)

        #============================================================

        what = data.findAll("div",attrs = {"class":"tag"})
        #print(what)
        try:
            tag = re.search('"haskey">(.*)</span>',str(what[i])).group(1)
        except:
            tag = "None"
        
        item[0]=title
        item[1]=name
        item[2]=district
        item[3]=unitprice
        item[4]=tag
        
        items[index]=item
        index+=1

        
        i += 1
    b+=1
```

```python
import csv
import codecs
with open("house.csv","w",newline = "",encoding ="utf-8-sig" ) as f:
    mywriter = csv.writer(f,dialect='excel')
    header = ["Tittle","Name","District","Unit Price","Remarks"]
    mywriter.writerow(header)
    mywriter.writerows(items)
    
import pandas as pd
df = pd.read_csv("house.csv")
df
```
